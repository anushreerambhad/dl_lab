```
import math
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-5,5,0.5)
x
```

```
def linear_activation_funct(x):
  ans=[]
  for array_element in x:
    ans.append(array_element)
  return ans
  
```

```
y = linear_activation_funct(x)
plt.plot(x,y)
plt.show()
```

```
def relu_activation_funct(x):
  ans=[]
  for array_element in x:
    if( array_element > 0):
      ans.append(array_element)
    else:
      ans.append(0)
  return ans
```

```
y1 = relu_activation_funct(x)
plt.plot(x,y1)
plt.show()
```

```
def sigmoid_activation_funct(x):
  ans=[]
  for array_element in x:
    ans.append(1/(1+np.exp(-array_element)))
    
  return ans


  
```

```
y2 = sigmoid_activation_funct(x)
plt.plot(x,y2)
plt.show()
```

```
#f1 = x**3 + x**2 + x + 2
def funct1_activation_funct(x):
  ans=[]
  for array_element in x:
    ans.append(array_element**3 + array_element**2 + array_element + 2)
    
  return ans
```

```
#tanh t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
def tanh_activation_funct(x):
  ans=[]
  for array_element in x:
    ans.append((np.exp(array_element)-np.exp(-array_element))/(np.exp(array_element)+np.exp(-array_element)))
    
  return ans
```

```
y4 = funct1_activation_funct(x)
plt.plot(x,y4)
plt.show()
```

```
def funct2_approximation(x):
  ans=[]
  for array_element in x:
    n1 = relu(-5*array_element -7.7)
    n2 = relu(-1.2*array_element -1.3)
    n3 = relu(1.2*array_element + 1)
    n4 = relu(1.2*array_element -.2)
    n5 = relu(2*array_element -1.1)
    n6 = relu(5*array_element - 5)
    z = -n1 -n2 -n3 +n4 +n5 +n6
    ans.append(z)
  return ans

def relu(x):
  return( np.maximum(x,0))
  
  
```

```
x = np.arange(-2,2,.2)
y5 = funct2_approximation(x)
plt.plot(x,y5)
plt.show()
```
